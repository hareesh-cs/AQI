```python
import pandas as pd 
import matplotlib.pyplot as plt
import pickle
import app
loaded_model=pickle.load(open('random_forest_regression_model.pkl', 'rb'))
```

```python
df=pd.read_csv('real_2014.csv')
df1=pd.read_csv('real_2014.csv')
mean = df1['PM 2.5'].mean(skipna=True)
df1=df1.replace(0,mean) 
df1=df1['PM 2.5'].fillna(mean) 
#df1=df1.dropna()
df1=df1.values.tolist()
my_prediction=loaded_model.predict(df.iloc[:,:-1].values)
my_prediction=my_prediction.tolist()
plt.plot(my_prediction,label="Predicted data")
plt.plot(df1,label="Actual data")
plt.xlabel('Day')
plt.ylabel('PM 2.5')
plt.legend(loc='upper right')
graph=plt.show()
```
